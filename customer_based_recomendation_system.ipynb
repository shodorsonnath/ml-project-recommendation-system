```python
def process_and_replace_orderid(file_path):

    chunk_size = 100000

    orderid_counts = {}

    for chunk in pd.read_csv(file_path, chunksize=chunk_size):
        chunk_orderid_counts = chunk['ORDERID'].value_counts()
        for orderid, count in chunk_orderid_counts.items():
            if orderid in orderid_counts:
                orderid_counts[orderid] += count
            else:
                orderid_counts[orderid] = count


    orderid_counts_df = pd.DataFrame(list(orderid_counts.items()), columns=['ORDERID', 'COUNT'])

    result_df = pd.DataFrame()

    for chunk in pd.read_csv(file_path, chunksize=chunk_size):
        chunk = chunk.merge(orderid_counts_df, on='ORDERID', how='left')
        chunk['ORDERID'] = chunk['COUNT']
        chunk.drop(columns=['COUNT'], inplace=True)
        result_df = pd.concat([result_df, chunk], ignore_index=True)

    result_df.to_csv(file_path, index=False)
    print(f'Processed file saved to: {file_path}')

```

# Prepare Data for LSTM model

### Don't need 'USERNAME_' and 'ITEMNAME' for the model as we have 'USERID' and 'ITEMID' for that

### Don't need 'TOTALPRICE' for the model as 'AMOUNT' and 'PRICE' will be enough

```python
df = combined_df.drop(columns=['USERNAME_','ITEMNAME','TOTALPRICE'])
df
```

## convert catagorical data into numarical data

```python
from sklearn.preprocessing import LabelEncoder

df['CATEGORY4'] = df['CATEGORY4'].astype(str).str.upper()
df['BRAND'] = df['BRAND'].astype(str).str.upper()
df['USERGENDER'] = df['USERGENDER'].astype(str).str.upper()
df['DISTRICT'] = df['DISTRICT'].astype(str).str.upper()

category_encoder = LabelEncoder()
df['CATEGORY'] = category_encoder.fit_transform(df['CATEGORY4'])

brand_encoder = LabelEncoder()
df['BRAND1'] = brand_encoder.fit_transform(df['BRAND'])

gender_encoder = LabelEncoder()
df['GENDER'] = gender_encoder.fit_transform(df['USERGENDER'])

district_encoder = LabelEncoder()
df['DISTRICT_CODE'] = district_encoder.fit_transform(df['DISTRICT'])

df.drop(columns=['CATEGORY4', 'BRAND', 'USERGENDER', 'DISTRICT'], inplace=True)

print(df.head())
```

## Test data

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding, Input
from sklearn.model_selection import train_test_split
```

```python
df = df.head(10000)
df
```

```python
df = df.sort_values(by='USERID')
df
```

```python
# Normalize numerical features
#numerical_features = []
#scaler = StandardScaler()
#df[numerical_features] = scaler.fit_transform(df[numerical_features])

numerical_features = ['ORDERID','USERID', 'AMOUNT', 'UNITPRICE', 'PRICE', 'AGE', 'CATEGORY', 'BRAND1', 'GENDER', 'DISTRICT_CODE']
scaler = MinMaxScaler()
df[numerical_features] = scaler.fit_transform(df[numerical_features])

df
```

```python
predicted_item_labels_y = df['ITEMCODE'].values


# Scale 'ITEMCODE' to range [0, 1]
scaler2 = MinMaxScaler()
df['ITEMCODE'] = scaler2.fit_transform(df[['ITEMCODE']])

df
```

```python
# Prepare input and output data

input_features = ['ORDERID','USERID', 'AMOUNT', 'UNITPRICE', 'PRICE', 'AGE', 'CATEGORY', 'BRAND1', 'GENDER', 'DISTRICT_CODE']
X = df[input_features]
y = df['ITEMCODE']
y
```

```python
sequence_length = 5

sequences = []
labels = []

# Generate sequences and corresponding labels
for i in range(len(X) - sequence_length):
    sequences.append(X.iloc[i:i + sequence_length].values)
    labels.append(y.iloc[i + sequence_length])

X_seq = np.array(sequences)
y_seq = np.array(labels)

```

```python
X_seq[0]
```

```python
y_seq[0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_seq, y_seq, test_size=0.2, random_state=42)

# Define the model
model = Sequential()
model.add(LSTM(100, input_shape=(X_seq.shape[1], X_seq.shape[2])))
model.add(Dense(1, activation='linear'))

# Compile the model
model.compile(loss='mse', optimizer='adam', metrics=['mae'])

# Print model summary
model.summary()

# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=32)
```

```python
predicted = model.predict(X_test)
```

```python
predicted_item_labels_X = scaler2.inverse_transform(predicted)
```

```python
predicted_item_labels_y = scaler2.inverse_transform(y_test.reshape(-1, 1))
predicted_item_labels_y
```

```python
# Evaluate the model
loss, accuracy = model.evaluate(predicted_item_labels_X, predicted_item_labels_y)
print(f'Test Loss: {loss}')
print(f'Test Accuracy: {accuracy}')
```
