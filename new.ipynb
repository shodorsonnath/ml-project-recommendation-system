# Prepare for model training

```python
df = combined_df[:100000].copy()
```

Split DATE_ into Year, Month, and Date

```python
df[["Year", "Month", "Date"]] = df['DATE_'].str.split("-", expand=True)
df
```

## Feature Engineering

**Convert USERBIRTHDATE to AGE**

```python
df["USERBIRTHDATE"] = pd.to_datetime(df["USERBIRTHDATE"])
df["AGE"] = (pd.to_datetime('today') - df["USERBIRTHDATE"]).dt.days
df
```

## Drop unnecessary columns

```python
df.drop(['ID', 'ORDERDETAILID', 'DATE_', 'USERNAME_', 'ITEMNAME',
         'NAMESURNAME', 'STATUS_', 'ITEMCODE', 'CATEGORY1', 'CATEGORY2',
         'CATEGORY3', 'REGION', 'CITY', 'TOWN', 'USERBIRTHDATE',
         "ADDRESSTEXT", "ADDRESSID", "rownum"], axis=1, inplace=True)
df
```

## Sort values

```python
df.sort_values(["USERID", "ORDERID", "Year", "Month", "Date"], inplace=True, ascending=True)
df
```

```python
df["Y_ITEMID"] = df["ITEMID"].shift(periods=-1)
df
```

```python
bool_mask_1 = (df["USERID"] == df['USERID'].shift(periods=-1))
bool_mask_2 = (df["ORDERID"] == df['ORDERID'].shift(periods=-1))
df = df[bool_mask_1 & bool_mask_2].copy()
df
```

## Encode categorical columns

```python
from sklearn.preprocessing import OrdinalEncoder
import pandas as pd

enc = OrdinalEncoder()
```

```python
df.loc[:, ["CATEGORY4", "BRAND", "USERGENDER", "DISTRICT"]] = enc.fit_transform(df[["CATEGORY4", "BRAND", "USERGENDER", "DISTRICT"]])
df
```

```python
df.dropna(inplace=True)
```

```python
df = df.astype('float')
df.info()
```

## Linear Regression

```python
from sklearn.model_selection import train_test_split
X  = df.drop("Y_ITEMID" , axis = 1)
y = df["Y_ITEMID"]

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor(max_depth = 4,n_estimators = 50, random_state=0)
regr.fit( X_train , y_train )
regr.score( X_test , y_test )
```
